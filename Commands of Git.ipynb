# **Essential Git Commands**

## Configuring Git


1. **Initializing a Repository**
   - `git init` to create a new repository.

2. **Checking the Git Version**
   - `git --version` to check if Git is installed and see the version.

3. **Checking the Status of Files**
   - `git status` to view modified files and their state in the repository.
   



4. **Adding Files to the Staging Area**
   - `git add <file_name>` to stage specific files.
   - `git add .` to stage all changes in the directory.

5. **Committing Changes**
   - `git commit -m "message"` to commit staged changes with a descriptive message.
    


6. **Viewing Commit History**
   - `git log` for a detailed history.
   - `git log --oneline` for a simplified view.



7. **Branching and Merging**
   - `git branch <branch_name>` to create a new branch.
   - `git checkout <branch_name>` or `git switch <branch_name>` to switch between branches.
   - `git merge <branch_name>` to merge a branch into the main branch.


8. **Git Diff**
   - `git diff` to see differences between your working directory and staged files.
    


9. **Stashing Changes**
   - `git stash` to temporarily save changes.
   - `git stash pop` to apply stashed changes.
   


10. **Git Rebase**
    - `git rebase <branch_name>` to rebase commits from one branch onto another.

11. **Pushing Changes to a Remote Repository**
    - `git push origin <branch_name>` to push changes to a remote repository.
    

12. **Pulling Changes from a Remote Repository**
    - `git pull origin <branch_name>` to update your local repository with remote changes.

13. **Creating and Managing Tags**
    - `git tag -a v1.0 -m "Version 1.0"` to create an annotated tag.
    - `git tag` to list all tags.

### **.gitignore**
   - Use a `.gitignore` file to exclude specific files or directories from being tracked by Git.  
   - Example:  
     ```
     *.log  
     node_modules/  
     .DS_Store
     ```

### **Handling Merge Conflicts**
   - When merging branches, conflicts may arise, and you must manually resolve them. Git will highlight conflicts, and you need to edit the file to decide which changes to keep. Example:
     - `git merge <branch_name>` to attempt merging and handle conflicts as they appear.
