## What is Git?
Imagine working on a large project with several team members. Everyone is making changes to the same files, and keeping track of who did what, and when, can quickly become overwhelming. Git helps solve this problem by tracking every change, so you can focus on coding without worrying about losing progress or overwriting someone else’s work.

## Definition?
Git is a distributed version control system designed to track changes in source code during software development. Created by Linus Torvalds in 2005, Git allows multiple developers to collaborate on projects by managing versions, branching, and merging code changes efficiently. Unlike older systems, Git lets developers work independently and then merge their changes seamlessly, making it an essential tool for modern software development.

##  What is Version Control System (VCS)?

A Version Control System (VCS) records changes to files or a set of files over time, so you can recall specific versions later. It enables multiple collaborators to work on a project simultaneously, track changes, and recover files if needed.

**Types of VCS**:

  **1.Local VCS**: Stores file changes on a local system.
  
  **2.Centralized VCS**: Allows multiple users to access a central server.
  
  **3.Distributed VCS**: Each collaborator has a full mirror of the project repository, offering greater fault tolerance.


## Why do we need Git? 

**1.Track Changes**: Git keeps a record of every change made to your code, allowing you to track and manage versions easily.

**2.Collaboration**: Multiple developers can work on the same project simultaneously without conflicts or losing work.

**3.Backup and Recovery**: Git provides a reliable way to back up your project, so you can recover your work if something goes wrong or files get lost.

## What is Github?

GitHub is a web-based platform that provides hosting for Git repositories. It allows developers to store, manage, and share their code online using Git. GitHub also offers additional features like collaboration tools, issue tracking, and project management, making it easier for teams to work together on projects. It’s widely used for open-source development and private projects, integrating with Git to enable seamless version control and code sharing.

## Difference between git and github

1.**Git** is a version control system used to track changes in code locally.     

2.**GitHub** is a cloud-based platform for hosting and sharing Git repositories online.    

3.**Git** works offline on your local machine, while **GitHub** requires an internet connection to access and collaborate on repositories.  

## Key Concepts of Git

**1. Repository (Repo)**: A directory that holds the project files and tracks their revision history.   
       - Local Repo: Stored on the user's machine.  
       - Remote Repo: Stored on an external server, e.g., GitHub, GitLab.   
    
**2. Commit**: A checkpoint in the project where changes are saved. Each commit has a unique identifier (SHA).
    
**3. Branch**: A parallel version of the repository, where independent work can be done.
    
**4. Working Directory**: The local files you are working on.
       
**5. Staging Area**: The space where changes are prepared before committing.
    
**6. SHA**: A unique identifier for every commit, represented as a 40-character string.

## How to Install Git?

### For Windows:

1. **Visit the Official Git Website**
   - Open your web browser and go to the official Git website: [https://git-scm.com](https://git-scm.com).
   - Click the **Download for Windows** button. It will automatically start downloading the Git installer for Windows.

2. **Run the Installer**
   - Once the Git installer `.exe` file is downloaded, find it in your Downloads folder and double-click to open it.
   
3. **Start the Installation Process**
   - When the Git Setup window opens, click **Next** to proceed through the setup screens.
   
4. **Choose Installation Location**
   - Select the folder where you want Git to be installed. By default, Git will be installed in the `C:\Program Files\Git` folder. Click **Next** to continue.

5. **Select Components**
   - In the "Select Components" window, you can choose additional options like a Git Bash shortcut or Git GUI. It’s usually fine to leave the default options selected. Click **Next** to continue.

6. **Choosing the Default Text Editor**
   - You’ll be prompted to choose the default text editor for Git. By default, it will use Vim, but you can change it to another editor, like Notepad++ or Visual Studio Code, if preferred. Select your preferred editor and click **Next**.

7. **Adjusting Path Environment**
   - In this step, select the option "Git from the command line and also from 3rd-party software". This is the recommended option for easy access to Git from both Git Bash and other terminal programs. Click **Next**.

8. **Choosing HTTPS Transport Backend**
   - Leave the default option "Use the OpenSSL library" selected and click **Next**.

9. **Completing the Installation**
    - The installer will now be ready to install Git. Click **Install** to start the installation process. Wait for the installation to complete.

10. **Launch Git Bash**
    - Once the installation is complete, you can choose to launch Git Bash immediately. Git Bash is a command-line interface that allows you to interact with Git. Check the box if you want to launch it now, or uncheck it and click **Finish**.
